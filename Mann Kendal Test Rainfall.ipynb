```python
import pandas as pd 
import numpy as np
import sklearn as sk
import matplotlib.pylab as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pymannkendall as mk
import statsmodels.api as sm
```

```python
data = pd.read_csv("E:\\DATA ANALYST\\rainfall.csv")
data.head()
plt.plot(data["Year"],data["annual"])
plt.xlabel("Year")
plt.ylabel("Annual (mm)")
plt.title("Rainfall Trends in Srikakulam")
plt.show()
```

```python
df = pd.read_csv("E:\\DATA ANALYST\\rainfall.csv")
df.head()
df.drop(['State','District','Year','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','annual','Mar-May (Summer)','Jun-Sep (Monsoon)','Oct-Dec (Fall)','Annuals'], axis='columns', inplace=True)
```

```python
Jan_Feb = mk.original_test(df)
```

```python
print(mk.seasonal_test(df,period=12))
```

```python
df.plot(figsize=(16, 7));
plt.show()
fig, ax = plt.subplots(figsize=(12, 8))
sm.graphics.tsa.plot_acf(df, lags=20, ax=ax);
plt.show()
```

```python
print(mk.sens_slope(df))
```
